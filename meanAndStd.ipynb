```python
import numpy as np
import matplotlib.pyplot as plt

# Example: 10 arrays with different lengths, for demonstration
# Replace these with your actual cost function arrays
cost_function_arrays = [
    np.array([10, 9.5, 9, 8.5, 8]),
    np.array([11, 10.5, 10, 9.5, 9]),
    np.array([9, 8.7, 8.5, 8, 7.8]),
    np.array([10.2, 9.8, 9.4, 9, 8.7]),
    np.array([10.5, 10, 9.6, 9.2, 8.9]),
    np.array([9.8, 9.4, 9, 8.7, 8.3]),
    np.array([10.1, 9.9, 9.5, 9.1, 8.8]),
    np.array([9.7, 9.3, 9, 8.6, 8.2]),
    np.array([10.3, 10, 9.6, 9.3, 9]),
    np.array([10.4, 10.1, 9.7, 9.4, 9.1])
]

# Convert list of arrays to a 2D numpy array
# Ensure all arrays are of the same length, or pad them to the same length if needed
max_length = max(len(arr) for arr in cost_function_arrays)
padded_arrays = np.array([np.pad(arr, (0, max_length - len(arr))) for arr in cost_function_arrays])

# Calculate mean and standard deviation along the rows
mean_values = np.mean(padded_arrays, axis=0)
std_deviation = np.std(padded_arrays, axis=0)
```

```python
mean_values
```

```python
# Generate iteration indices (assuming equal length for simplicity)
iterations = np.arange(1, max_length + 1)

# Plot mean and standard deviation
plt.figure(figsize=(10, 6))
plt.plot(iterations, mean_values, label='Mean Cost Function Value', color='blue')
plt.fill_between(iterations, mean_values - std_deviation, mean_values + std_deviation, color='blue', alpha=0.2)
plt.xlabel('Iteration')
plt.ylabel('Cost Function Value')
plt.title('Mean and Standard Deviation of Cost Function Values')
plt.legend()
plt.grid(True)
plt.show()
```
